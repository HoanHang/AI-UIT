```python
# Use a pipeline as a high-level helper
from transformers import pipeline
```

```python
pipe = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B", trust_remote_code=True)
```

```python
messages = [
    {"role": "user", "content": "1+1=?, give a code python to solve this problem"},
]
```

```python
a = pipe(messages)
```

```python
print(a[0]['generated_text'][1]['content'])
```

Yêu cầu 1: Viết một prompt yêu cầu LLM định nghĩa “học sâu (deep learning)” dành cho người chưa có nền tảng kỹ thuật.

```python
messages = [{
    "role": "user",
    "content": "Giải thích khái niệm 'học sâu (deep learning)' bằng tiếng Việt đơn giản, "
               "dùng ví dụ thực tế và so sánh dễ hiểu. Tránh dùng thuật ngữ kỹ thuật, "
               "hướng tới người chưa có kiến thức AI."
}]
a = pipe(messages)
print(a[0]['generated_text'][1]['content'])
```

Yêu cầu 2: Viết prompt yêu cầu LLM liệt kê 5 ứng dụng của AI trong đời sống hằng ngày, trình bày dưới dạng bảng.

```python
messages = [{
    "role": "user",
    "content": "Liệt kê 5 ứng dụng phổ biến của AI trong đời sống hàng ngày. "
               "Trình bày dạng bảng gồm: Tên ứng dụng - Lĩnh vực - Mô tả ngắn (không quá 15 từ). "
               "Ưu tiên ví dụ tại Việt Nam."
}]
a = pipe(messages)
print(a[0]['generated_text'][1]['content'])
```

Yêu cầu 3: Viết prompt yêu cầu LLM đóng vai chuyên gia hướng nghiệp tư vấn cho sinh viên ngành Khoa học máy tính năm cuối.


```python
messages = [{
    "role": "user",
    "content": "Bạn đang đóng vai chuyên gia hướng nghiệp CNTT. Hãy tư vấn cho sinh viên Khoa học Máy tính năm cuối: "
               "1) 3 lộ trình nghề nghiệp hot tại Việt Nam "
               "2) Kỹ năng cần bổ sung cho mỗi lộ trình "
               "3) Sai lầm cần tránh khi mới ra trường. "
               "Dùng giọng văn thân thiện, định hướng hành động cụ thể."
}]
a = pipe(messages)
print(a[0]['generated_text'][1]['content'])
```

Yêu cầu 4: Viết prompt yêu cầu LLM tạo đoạn mã Python mô phỏng mạng nơ-ron đơn giản, gồm 1 layer input có 3 nodes, 1 hidden layer có 5 nodes và 1 ouput layers có 3 nodes.

```python
messages = [{
    "role": "user",
    "content": "Viết mã Python tạo mạng neural đơn giản dùng PyTorch với: "
               "- Lớp input: 3 nodes "
               "- Lớp ẩn: 5 nodes (kèm kích hoạt ReLU) "
               "- Lớp output: 3 nodes "
               "Yêu cầu: "
               "1) Random khởi tạo trọng số "
               "2) Tính forward pass "
               "3) Demo với input mẫu [1.0, 2.0, 3.0] "
               "4) Giải thích ngắn từng bước trong code."
}]
a = pipe(messages)
print(a[0]['generated_text'][1]['content'])
```

Yêu cầu 5: Viết prompt yêu cầu LLM gợi ý đề tài khóa luận AI kèm mục tiêu, phương pháp, ứng dụng

```python
messages = [{
    "role": "user",
    "content": "Gợi ý 3 đề tài khóa luận AI cho sinh viên CNTT, mỗi đề tài trình bày theo cấu trúc: "
               "1) Tên đề tài "
               "2) Mục tiêu nghiên cứu (tối đa 2 gạch đầu dòng) "
               "3) Phương pháp tiếp cận (liệt kê ngắn) "
               "4) Ứng dụng thực tế. "
               "Ưu tiên đề tài khả thi trong 6 tháng, dữ liệu công khai."
}]
a = pipe(messages)
print(a[0]['generated_text'][1]['content'])
```
